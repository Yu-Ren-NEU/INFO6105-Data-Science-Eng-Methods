```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
import pymc3 as pm
from pymc3.ode import DifferentialEquation
import theano
from pymc3 import Model, Lognormal
```

## SIR MODEL

```python
def SIR(y, t, p):
    di = p[0] * (1. - y[0] - y[1] - p[1]/p[0]) * y[0]
    dr = p[1] * y[0]
    return [di, dr]
```

```python
def susceptible(i, r):
    s = 1 - i - r
    return s
```

```python
times = np.arange(0, 5, 0.25)
```

```python
beta, gamma = 4.0, 1.0
```

```python
# Compute true curves
y = odeint(SIR, t=times, y0=[0.01, 0.0], args=((beta, gamma),), rtol=1e-8)
```

```python
yobs = np.random.lognormal(mean=np.log(y[1::]), sigma=[0.1, 0.1])
```

```python
plt.plot(yobs[:,0], 'b', label="Infective")
plt.plot(yobs[:,1], '--g', label="Recovered/Dead")
# plt.plot(susceptible(yobs[:,0], yobs[:,1]), ':r', label="Susceptible")

plt.xlabel('time')
plt.ylabel('rate')

plt.title("SIR Model")

plt.legend()
plt.show()
```

## Import Data

```python
China = pd.read_csv('Train_data.csv')
Wuhan=China[China['cityEnglishName']=='Wuhan']
Wuhan['city_recoveredCount'] = Wuhan['city_curedCount'] + Wuhan['city_deadCount']
drop_cols = ['provinceName', 'provinceEnglishName', 'province_zipCode', 'cityName', 'city_zipCode', 'province_suspectedCount', 'city_deadCount', 'province_deadCount','province_confirmedCount','province_curedCount','city_suspectedCount', 'city_curedCount']
Wuhan.drop(drop_cols, axis=1, inplace=True)
Wuhan.sort_values(by='updateTime', inplace=True)
Wuhan.reset_index(drop=True, inplace=True)
Wuhan
```

## Compute beta and gamma

```python
Wuhan_extension = Wuhan.sort_values(by='updateTime',ascending=False)
Wuhan_extension['city_recoveredCount'] += (max(Wuhan_extension['city_recoveredCount'])-Wuhan_extension['city_recoveredCount'])*2
Wuhan_extension.reset_index(drop=True, inplace=True)
Wuhan_extension
```

```python
Wuhan_final = pd.concat([Wuhan, Wuhan_extension])
Wuhan_final.reset_index(drop=True, inplace=True)
Wuhan_final.drop(['updateTime', 'cityEnglishName'], axis=1, inplace=True)
Wuhan_final = Wuhan_final[12::30]
Wuhan_final
```

```python
sir_model = DifferentialEquation(
    func = SIR,
    times = np.arange(0.25, 5, 0.2),
    n_states = 2,
    n_theta = 2,
    t0 = 0,
)
```

```python
with pm.Model() as covid19:
    sig = pm.HalfCauchy('sig', 1, shape=2)
    R0 = pm.Bound(pm.Normal, lower=1)('R0', 2, 3)
    gama = pm.Lognormal('gama', pm.math.log(2), 2)
    beta = pm.Deterministic('beta', gama*R0)
    sir_curves = sir_model(y0=[0.01, 0.0], theta=[beta, gama])
    Y = pm.Lognormal('Y', mu=pm.math.log(sir_curves), sd=sig, observed=Wuhan_final)
    prior = pm.sample_prior_predictive()
    trace = pm.sample(2000, tune=1000, target_accept=0.9, cores=1)
```

```python
pm.traceplot(trace)
plt.show()
```

```python
pm.plot_posterior(trace)
plt.show()
```
