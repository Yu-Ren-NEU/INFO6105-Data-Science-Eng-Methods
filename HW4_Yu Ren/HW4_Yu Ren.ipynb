```python
# Required Packages 

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import svm  
from sklearn.datasets import load_iris
```

```python
# import iris data to model Svm classifier (3 points)

iris_dataset = load_iris()
```

```python
# Using the DESCR key over the iris_dataset to describ the dataset (3 points)

print(iris_dataset['DESCR'])
```

```python
# To get the iris features and the target classes (3 points)

print("Iris features are:", iris_dataset.feature_names)
print("Iris target classes are:", iris_dataset.target_names)
```

```python
# To check the target data (3 points)

print(iris_dataset.target)
```

```python
X = iris_dataset.data[:, :2]
y = iris_dataset.target
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.xlabel('Sepal Lenght')
plt.ylabel('Sepal Width')
plt.title('The relationship between sepal and target classes')
plt.show()
```

```python
X = iris_dataset.data[:, 2:]
y = iris_dataset.target
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.xlabel('Petal Lenght')
plt.ylabel('Petal Width')
plt.title('The relationship between petal and target classes')
plt.show()
```

```python
X = iris_dataset.data[:, :2]
y = iris_dataset.target
C = 1.0

lin_svc = svm.SVC(kernel='linear', C=C).fit(X, y)
rbf_svc = svm.SVC(kernel='rbf', gamma='auto', C=C).fit(X, y)
poly_svc = svm.SVC(kernel='poly', gamma='auto', degree=3, C=C).fit(X, y)
```

```python
x_min = X[:, 0].min() - 0.2
x_max = X[:, 0].max() + 0.2
y_min = X[:, 1].min() - 0.2
y_max = X[:, 1].max() + 0.2
xx, yy = np.meshgrid(np.arange(x_min, x_max, .02), np.arange(y_min, y_max, .02))

titles = ['SVC with linear kernel', 'SVC with RBF kernel', 'SVC with polynomial (degree=3) kernel']

for i, classifier in enumerate((lin_svc, rbf_svc, poly_svc)):
    plt.subplot(2, 2, i + 1)
    plt.subplots_adjust(wspace=0.5, hspace=1)
    
    Z = classifier.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    
    plt.contourf(xx, yy, Z, alpha=0.8)
    
    plt.scatter(X[:, 0], X[:, 1], c=y)
    plt.xlabel('Sepal length')
    plt.ylabel('Sepal width')
    plt.xlim(xx.min(), xx.max())
    plt.ylim(yy.min(), yy.max())
    plt.title(titles[i])
    
plt.show()
```

```python
print('Accuracy of Linear SVC: {:.5f}'.format(lin_svc.score(X, y)))
print('Accuracy of RBF SVC: {:.5f}'.format(rbf_svc.score(X, y)))
print('Accuracy of Polynomial SVC: {:.5f}'.format(poly_svc.score(X, y)))
```
