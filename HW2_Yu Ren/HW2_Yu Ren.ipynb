```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
t = np.arange(0., 5., 0.2)
plt.plot(t, t, '--g')
plt.plot(t, t**2, 'bs')
plt.plot(t, t**3, 'r^')
plt.show()
```

```python
x1 = [10,20,30]
y1 = [20,40,10]

x2 = [10,20,30]
y2 = [40,10,30]

plt.xlabel('x')
plt.ylabel('y')

plt.plot(x1, y1, ':b', label='line1-dotted')
plt.plot(x2, y2, '--r', label='line2-dashed')

plt.title("Plot with two or more lines with different styles")

plt.legend()

plt.show()
```

```python
x = np.random.randn(50)
y = np.random.randn(50)

plt.scatter(x, y, s=100, facecolors='none', edgecolor='r');
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'June', 'July', 'Aug', 'Sept', 'Oct', 'Nov', 'Dec']
precip = [0.70, 0.75, 1.85, 2.93, 3.05, 2.02, 1.93, 1.62, 1.84, 1.31, 1.39, 0.84]
plt.plot(months, precip);

# set plot size for the plot
plt.rcParams["figure.figsize"] = (8, 8)

plt.show()
```

```python
plt.plot(months, precip);
# set plot size for the plot
plt.rcParams["figure.figsize"] = (8, 8)
plt.title("Average Monthly Precipitation")
plt.xlabel("Month")
plt.ylabel("Precipitation (inches)")

plt.show()
```

```python
plt.bar(months, precip, color="green")

# set plot size for the plot
plt.rcParams["figure.figsize"] = (8, 8)
plt.title("Average Monthly Precipitation")
plt.xlabel("Month")
plt.ylabel("Precipitation (inches)")

plt.show()
```
